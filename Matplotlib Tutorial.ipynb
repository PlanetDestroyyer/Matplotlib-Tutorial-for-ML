Bar Plots

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languagues = ['English','Latin',"German",'French','Hindi']
people = [100,50,16,60,200]
ax.bar(languagues,people)
plt.xlabel("Languages")
plt.ylabel("People")
plt.show()
```

Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languagues = ['English','Latin',"German",'French','Hindi']
people = [100,50,16,60,200]
ax.pie(people,labels=languagues,autopct='%1.1f%%')
plt.show()
```

Scatter Plot



```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y)
ax.scatter(x,z,color='r')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z =20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c='red')
plt.show()
```
